## **i. Calculating Measures of Central Tendency:**
These measures provide a single value that represents the central or typical value of a dataset. The mean is the arithmetic average of all the values in the dataset, the median is the middle value when the data is arranged in order, and the mode is the most frequently occurring value in the dataset.

## **Calculating Measures of Variability:**
These measures provide a sense of the spread or dispersion of a dataset. They include the range, which is the difference between the maximum and minimum values in the dataset, the standard deviation, which measures how much the data deviates from the mean, and the variance, which is the average of the squared deviations from the mean.

```python
#@title
import pandas as pd
import scipy.stats as stats

# Load the tips dataset
df = sns.load_dataset('tips')
# Select only numeric columns
df = df.select_dtypes(include='number')

# Compute measures of variability
measures_of_variability = {
'range': df.max() - df.min(),  # Difference between the maximum and minimum values
    'variance': df.var(),  # Average of the squared differences from the mean
    'std_dev': df.std(),  # Standard deviation, which is the square root of variance
    'coeff_variation': df.std() / df.mean(),  # Coefficient of variation, standard deviation divided by the mean
    'IQR': df.quantile(0.75) - df.quantile(0.25),  # Interquartile range, difference between the 75th and 25th percentiles
    'skewness': df.skew(),  # Skewness, a measure of the asymmetry of the distribution
    'kurtosis': df.kurt(),  # Kurtosis, a measure of the "tailedness" of the distribution
    # 'median_absolute_deviation': df.mad(), # Deprecated since Pandas 1.5, replaced with a custom MAD function
    'sem': df.sem(),  # Standard error of the mean
}

# Convert the dictionary to a pandas dataframe and transpose it to have variables as columns
df_variability = pd.DataFrame.from_dict(measures_of_variability).T

# Print the resulting datapd.DataFrame.from_dict(measures_of_variability, orient='columns')
df_variability
```

## **Creating Frequency Tables:**
These tables are used to summarize **categorical variables** by counting the number of times each category occurs in the dataset. They provide a way to describe the distribution of the categories in the data.

```python
#@title Frequency Tables for categorical variables
import pandas as pd

# Load the tips dataset
df = sns.load_dataset('tips')

import pandas as pd
import seaborn as sns

# Load dataset
df = sns.load_dataset('tips')

# Create frequency tables for all categorical variables
freq_tables = {
    'sex': df['sex'].value_counts(),
    'smoker': df['smoker'].value_counts(),
    'day': df['day'].value_counts(),
    'time': df['time'].value_counts()
}

# Print each frequency table
for variable, freq_table in freq_tables.items():
    print(f"\nFrequency table for {variable}:")
    print(freq_table)
```

### **i. Create scatter plots with trend line**


A scatter plot with a trend line is a plot where each data point is represented by a dot, and a line is drawn through the points to show the general trend of the data. The trend line is calculated using regression analysis and can help identify any correlation between the two variables.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
df = sns.load_dataset('iris')
display(df)
df.describe()
```

```python
sns.lmplot(x='sepal_length', y='petal_length', data=df)
# sns.lmplot(x='sepal_width', y='petal_width', data=df)
```

This code creates a scatter plot with a trend line showing the relationship between sepal length and petal length. From the plot, we can see that there is a **strong positive correlation between the two length variables**, which means that as the sepal length increases, the petal length tends to increase as well. The trend line confirms this relationship, with a positive slope indicating a positive correlation.

Moreover, we can see some variability in the data points around the trend line, which suggests that there may be some other factors affecting the relationship between sepal length and petal length.

 **A negative correlation is true for width variables.**

```python
# Calculate correlation coefficients
corr_sepal_width_petal_width = df['sepal_width'].corr(df['petal_width'])
corr_sepal_length_petal_length = df['sepal_length'].corr(df['petal_length'])

print("Correlation between Sepal Width and Petal Width:", corr_sepal_width_petal_width.round(3))
print("Correlation between Sepal Length and Petal Length:", corr_sepal_length_petal_length.round(3))
```

The correlation coefficients show an interesting negative corelation (-0.366) between the widths, while the lengths almost have a 90% correlation!

To further investigate the relationship between sepal length and petal length with respect to the different iris species, add color-coding to the scatter plot using the hue parameter.

In a pair plot, multiple scatter plots are displayed in a grid, with each scatter plot showing the relationship between two variables. Additionally, a regression line is added to each scatter plot to help visualize the strength and direction of the relationship. The data points are also color-coded by species to help distinguish between groups or categories.

**Pros**:

    • Allows for quick and easy visualization of multiple relationships between variables.

    • Color-coding by species helps to identify patterns or differences between groups.

    • Regression lines help to identify the direction and strength of the relationships between variables.

**Cons**:

    • Pair plots can become cluttered and difficult to read if there are too many variables being plotted.

    • Interpretation of the regression lines can be challenging for those without a statistical background.
    
    • Computationally Intensive: May not be suitable for large datasets or datasets with many categorical variables.

## Notches best practices:

Choosing between a 95% and 99.5% confidence interval for notches in a box plot depends on the desired level of certainty in the difference between medians. A 95% confidence interval may be sufficient in most cases, but if there is a need for higher certainty, such as in medical or financial research, a 99.5% confidence interval may be more appropriate. For example, a 99.5% confidence interval could be used in clinical trials to compare the effectiveness of two treatments.

## Whisker length best practices:

• Use the default whisker length **1.5 times the IQR** unless there is a good reason to change it.

• If there are a lot of data points and outliers are not of primary interest, consider reducing the whisker length to make the box plot more readable.

• If outliers are of interest, consider increasing the whisker length to show more of them.

• Consider including individual data points, such as by using Seaborn's `swarmplot`, to give a more complete picture of the distribution.

• Avoid truncating the whiskers, as this can be misleading and obscure important features of the data.

```python
# @title iii. Swarm plots already have jitter, but see this bubble stripplot from first principals

# Map species to numerical values
species_dict = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
df['species_num'] = df['species'].map(species_dict)

# Define a function to set point sizes based on a third variable
def set_point_sizes(data, x_col, y_col, size_col, ax, size_range=(10, 100)):
    size_scale = lambda x: ((x - data[size_col].min())
                            / (data[size_col].max() - data[size_col].min()))
    point_sizes = size_scale(data[size_col]) * (size_range[1] - size_range[0]) + size_range[0]
    ax.scatter(data[x_col], data[y_col], s=point_sizes, alpha=0.5)

# Create a stripplot with point size based on sepal length
fig, ax = plt.subplots()
set_point_sizes(df, 'species_num', 'petal_width', 'sepal_length', ax)
ax.set_xlabel('Species')
ax.set_ylabel('Petal Width')
ax.set_title('Bubble Plot with Stripplot')
plt.show()
```

## **i.   Correlation matrix with different correlation coefficients -- Pearson, Spearman and Kendall**

**Pearson** is most **commonly** used.

**Spearman and Kendall** correlation coefficients are **non-parametric** measures -- more robust to non-normality and non-linear relationships. **Spearman** correlation coefficient is suitable when the data is **ordinal**,
whereas **Kendall** correlation coefficient is suitable for **both** ordinal and continuous data, especially larger datasets.

The vmin and vmax parameters are set to -1 and 1, respectively, to ensure that the color scale ranges from -1 (perfect negative correlation) to 1 (perfect positive correlation).



```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Load the iris dataset
iris = sns.load_dataset("iris")
df = iris.copy()

display(df.head())

# First convert all variables (actually only species was necessaru)
for col in df.columns:
    df[col] = LabelEncoder().fit_transform(df[col])

# Compute the correlation matrix for the iris dataset
corr_pearson = df.corr(method='pearson')
corr_spearman = df.corr(method='spearman')
corr_kendall = df.corr(method='kendall')

# Create a figure with three subplots for each correlation matrix
fig, axs = plt.subplots(ncols=3, figsize=(19,5))

# Easy upper triangle mask
up_mask = np.triu(corr_pearson)

# Plot the Pearson correlation matrix
sns.heatmap(corr_pearson, cmap='coolwarm', annot=True, cbar=False, square=True, mask=up_mask, ax=axs[0])
axs[0].set_title('Pearson Correlation Matrix')

# Plot the Spearman correlation matrix
sns.heatmap(corr_spearman, cmap='coolwarm', annot=True, cbar=False, square=True, mask=up_mask, ax=axs[1])
axs[1].set_title('Spearman Correlation Matrix')

# Plot the Kendall correlation matrix
sns.heatmap(corr_kendall, cmap='coolwarm', annot=True, cbar=False, square=True, mask=up_mask, ax=axs[2])
axs[2].set_title('Kendall Correlation Matrix')

# Show the tight plots
plt.tight_layout()
plt.show()
```

*   Pearson correlation matrix shows strong positive correlations between petal length and petal width.

*   Spearman correlation matrix shows a similar pattern of correlations, but with weak correlation between sepal length and sepal width.

*   Kendall correlation matrix, which is less sensitive to outliers, shows weaker correlations overall compared to the other two matrices.

Summary on correlation coefficients:

*   **Pearson** correlation coefficient: Use when both variables are **normally distributed** and/or have a **linear** relationship. It is sensitive to outliers and assumes that the data is **homoscedastic** -- same variance across all levels of the variable. In other words, the variability of the data points is similar across the range of the variable being measured. This assumption is important for the interpretation of the results of the Pearson correlation coefficient, which assumes homoscedasticity of the data.

*   **Spearman** correlation coefficient: Use when the relationship between the two variables is **monotonic** (i.e., consistently increasing or decreasing) but not necessarily linear, and when there may be outliers in the data. It is a **non-parametric measure**, so it does not assume a particular distribution of the data.

*   **Kendall** correlation coefficient: Use when the relationship between the two variables is **monotonic** and when there may be outliers in the data. It is also a **non-parametric measure**, but it is **less sensitive to outliers** than the Spearman coefficient.

## **Multicollinearity: Guidelines and best practices**

*   **Multicollinearity** is when predictor variables in a regression or ML model are highly correlated.
*   It can lead to unstable and unreliable model coefficients and make it difficult to interpret results.
*   Guidelines for addressing multicollinearity include examining correlation matrices and **removing highly correlated variables** or using techniques like:

   *   Remove correlated variables
   *   Scaling or transformation
   *   Use PCA or other subspace methods
   *   Use ridge regression
   *   Cluster variables
   *   Collect more data

**Correlation coefficient	Interpretation:**
*   0.05 - 0.3 	Weak correlation (sometimes 0.05 - 0.5)
*   0.3 - 0.7	Moderate correlation (sometimes 0.5 - 0.7)
*   0.7+        Strong correlation (typical)

**Significance level**, often denoted as alpha (α), is the probability of rejecting the null hypothesis when it is actually true. In other words, it is the probability of making a Type I error, which is the rejection of a true null hypothesis. It is a pre-determined threshold that is used to determine whether the observed results are statistically significant. The significance level is typically set at 0.05, which means that there is a 5% chance of rejecting the null hypothesis when it is actually true.

## Reminder of Significance Level:

**Significance level**, often denoted as alpha (α), is the probability of rejecting the null hypothesis when it is actually true. In other words, it is the probability of making a Type I error, which is the rejection of a true null hypothesis. It is a pre-determined threshold that is used to determine whether the observed results are statistically significant. The **significance level** is typically set at **0.05**, which means that there is a **5% chance** of rejecting the **null hypothesis** when it is actually true. Therefore, the **p-value** is a measure of the **strength of evidence against the null hypothesis** in a statistical hypothesis test.

An **hypothesis** is an educated guess that can be tested with observations and falsified if it really is false. You **cannot prove** conclusively that most **hypotheses** are **true** because it's generally impossible to examine all possible cases for exceptions that would disprove them.
### Therefore, a standard approach is to disprove a null hypothesis rather than proving an alternative hypothesis.

```python
#@title Hypothesis Testing Intro: ANOVA, t-tests or chi-square tests
import seaborn as sns
import pandas as pd
import scipy.stats as stats

# Load the tips dataset
df = sns.load_dataset('tips')

# Perform a t-test to compare the average total bill between smokers and non-smokers
smokers = df[df['smoker'] == 'Yes']['total_bill']
non_smokers = df[df['smoker'] == 'No']['total_bill']
ttest_result = stats.ttest_ind(smokers, non_smokers)
print('T-Test Results:')
print('-------------------------')
print('t-statistic:', ttest_result[0])
print('p-value:', ttest_result[1])
print('-------------------------')

# Perform an ANOVA to compare the average total bill among different day of the week
days = ['Thur', 'Fri', 'Sat', 'Sun']
f, p = stats.f_oneway(df[df['day'] == 'Thur']['total_bill'],
                      df[df['day'] == 'Fri']['total_bill'],
                      df[df['day'] == 'Sat']['total_bill'],
                      df[df['day'] == 'Sun']['total_bill'])
print('ANOVA Results:')
print('-------------------------')
print('F-value:', f)
print('p-value:', p)
print('-------------------------')

# Perform a chi-square test to examine the association between sex and smoker status.
crosstab = pd.crosstab(df['sex'], df['smoker'])
chi2_result = stats.chi2_contingency(crosstab)
print('Chi-Square Test Results:')
print('-------------------------')
print('Chi-Square Statistic:', chi2_result[0])
print('p-value:', chi2_result[1])
print('-------------------------')
```

```python
#@title Hypothetical Example: Comparing Total Bill Before and After a Promotional Event
from scipy.stats import ttest_rel
import numpy as np

# Create a hypothetical dataset
np.random.seed(42)
n_customers = 30

# Generate random total bills before and after a promotion
total_bill_before = np.random.normal(loc=30, scale=5, size=n_customers)  # Before promotion
total_bill_after = np.random.normal(loc=35, scale=5, size=n_customers)  # After promotion

# Create DataFrame
data = pd.DataFrame({
    'customer_id': range(1, n_customers + 1),
    'total_bill_before': total_bill_before,
    'total_bill_after': total_bill_after
})

# Perform the paired t-test
ttest_result = ttest_rel(data['total_bill_before'], data['total_bill_after'])
print('Paired T-Test Results:')
print('-------------------------')
print('t-statistic:', ttest_result.statistic)
print('p-value:', ttest_result.pvalue)
print('-------------------------')

# Interpretation
if ttest_result.pvalue < 0.05:
    print("The test is significant: There is a statistically significant difference in total bills before and after the promotion.")
else:
    print("The test is not significant: There is no statistically significant difference in total bills before and after the promotion.")
```

## How to:


**Interpret a T-Test**

*   **T-statistic**: The t-statistic measures the difference between the **means** of **two groups**, divided by the variation within the groups. A **higher** absolute **t-value** indicates a **larger difference** between the two means.
*   **P-value:** The p-value is the probability of obtaining the observed results by chance, assuming that the null hypothesis is true. A p-value less than the significance level (often **p < 0.05**) suggests that the difference between the two means is **statistically significant**. More extreme versions can also be defined such as highly significant (often p < 0.005)

**Interpret ANOVA**

*   **F-value:** The F-value is the ratio of the variation *between* groups to the variation *within* groups. A higher F-value indicates a larger difference between the means of the groups being compared.
    *   High F-value: Indicates that the variation between the group means is much larger than the variation within the groups, suggesting significant differences between the group means.
    *   Low F-value: Indicates that the variation between the group means is similar to the variation within the groups, suggesting no significant differences between the group means.
*   **P-value:** The p-value is the probability of obtaining the observed results by chance, assuming that the null hypothesis is true. A p-value less than the significance level (p < 0.05) suggests that at least one of the means being compared is statistically different from the others.

**Interpret Chi-Square Test**

*   **Chi-square statistic:** The chi-square statistic measures the difference between the observed frequencies and the expected frequencies under the null hypothesis. A higher chi-square statistic indicates a larger difference between the observed and expected frequencies.
*   **P-value:** The p-value is the probability of obtaining the observed results by chance, assuming that the null hypothesis is true. A p-value less than the significance level (often 0.05) suggests that the two categorical variables are not independent and there is a significant association between them.

```python
#@title ANOVA... When is the F-statistic too big?

from scipy.stats import f

# Define degrees of freedom
df_between = 3  # Example: number of groups - 1
df_within = 116  # Example: total number of observations - number of groups

# Define significance level
alpha = 0.05

# Calculate the critical F-value
critical_f_value = f.ppf(1 - alpha, df_between, df_within)
print(f'Critical F-Value at alpha={alpha} with df_between={df_between} and df_within={df_within}: {critical_f_value}')

# Example F-value to compare
f_value = 5.0

# Compare F-value with critical value
if f_value > critical_f_value:
    print("Reject the null hypothesis: significant differences among group means (F is larger than critical F by):", f_value-critical_f_value)
else:
    print("Fail to reject the null hypothesis: no significant differences among group means (F is less than or equal to the critical F):", f_value-critical_f_value)

```

```python
#@title Chi-Square... A better example

from scipy.stats import chi2_contingency, chi2

# Create the contingency table
crosstab = pd.DataFrame({
    'Smoker (Yes)': [30, 20],
    'Smoker (No)': [40, 30]
}, index=['Male', 'Female'])

# Perform the Chi-Square Test
chi2_stat, p_value, dof, expected = chi2_contingency(crosstab)

# Output the Chi-Square Test results
print(f'Chi-Square Statistic: {chi2_stat}')
print(f'p-value: {p_value}')
print(f'Degrees of Freedom: {dof}')
print(f'Expected Frequencies:\n{expected}')

# Calculate the critical value for the Chi-Square Distribution
alpha = 0.05  # Significance level
critical_value = chi2.ppf(1 - alpha, df=dof)
print(f'Critical Value at alpha={alpha} and df={dof}: {critical_value}')

# Compare the Chi-Square Statistic with the Critical Value
if chi2_stat > critical_value:
    print("Reject the null hypothesis: significant association between the variables.")
else:
    print("Fail to reject the null hypothesis: no significant association between the variables.")
```

```python
#@title Chi-Square... A strong association example
from scipy.stats import chi2_contingency, chi2

# Create the contingency table with a strong but realistic association (roflcopter)
crosstab = pd.DataFrame({
    'Tea': [60, 35, 20],
    'Coffee': [20, 45, 70]
}, index=['High School', 'Undergraduate', 'Graduate'])

# Perform the Chi-Square Test
chi2_stat, p_value, dof, expected = chi2_contingency(crosstab)

# Output the Chi-Square Test results
print(f'Chi-Square Statistic: {chi2_stat:.2f}')
print(f'p-value: {p_value:.4f}')
print(f'Degrees of Freedom: {dof}')
print(f'Expected Frequencies:\n{expected}')

# Calculate the critical value for the Chi-Square Distribution
alpha = 0.05  # Significance level
critical_value = chi2.ppf(1 - alpha, df=dof)
print(f'Critical Value at alpha={alpha} and df={dof}: {critical_value:.2f}')

# Compare the Chi-Square Statistic with the Critical Value
if chi2_stat > critical_value:
    print("Reject the null hypothesis: significant association between education level and beverage preference.")
else:
    print("Fail to reject the null hypothesis: no significant association between education level and beverage preference.")

```

### What is R-Square?
Remember Correlation Coefficients that indicates the strength of a linear relationship between two variables.? Well R-Square is different.

**R-Squared** represents the proportion of **variance explained** by the **model**, which can be influenced by the number of **predictors** and model **complexity**.

**Adjusted R-Squared** should be used when considering many predictors. Best practices here is to use it when > 10 variables (predictors) are used in regression.

### Guidelines.. why different :'(
*   **Low explanatory power**: 0.0 - 0.3
*   **Moderate explanatory power:** 0.3 - 0.6
*   **High explanatory power:** 0.6 and above


```python
#@title Regression Analysis: Predicting Total Bill
import statsmodels.api as sm

# Load the tips dataset
df = sns.load_dataset('tips')

# Create a new column for tip percentage
df['tip_pct'] = 100 * df['tip'] / df['total_bill']

# Use total bill and size as predictor variables to predict tip percentage
X = df[['total_bill', 'size']]
y = df['tip_pct']

# Add a constant term to the predictor variables for regression analysis
X = sm.add_constant(X)

# Fit the OLS regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())

# Interpretation of the results
print('Interpretation of the results:')
print('------------------------------------')
print('R-squared:', model.rsquared)
print('------------------------------------')
print('The R-squared value of', round(model.rsquared, 2), 'indicates that the model explains', round(model.rsquared * 100, 2), '% of the variation in the tip percentage.')
print('------------------------------------')
print('The coefficient for the total_bill variable is', round(model.params[1], 2), ', indicating that for every additional dollar spent on the total bill, the tip percentage is expected to increase by', round(model.params[1], 2), '%.')
print('------------------------------------')
print('The coefficient for the size variable is', round(model.params[2], 2), ', indicating that for every additional person in the group, the tip percentage is expected to decrease by', round(-model.params[2], 2), '%.')
print('------------------------------------')
print('The p-values for both predictor variables are less than the significance level of 0.05, indicating that both variables are statistically significant in predicting the tip percentage.')
```

### Categorical encoding advice
If you are using `pd.get_dummies` instead of label encoder... we prefer `drop_first=True`... why:

Let `quality` be a column with categorical levels "low", "medium" and "high":

One-hot encoding (`pd.get_dummies`) produces `quality_medium` and `quality_high` columns.

The reference category ("low") is clearly omitted.
The coefficients for quality_medium and quality_high represent the difference in the dependent variable compared to the "low" category. This defeats the multicollinearity problem caused by `drop_first=False`
